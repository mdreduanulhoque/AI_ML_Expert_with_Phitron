### Creating Class

```python
class phone:
    catagory = "Electronics"
    current_battery = 100
    # Constructor
    def __init__(self,model,camera,battery):
        self.model = model
        self.camera = camera
        self.battery = battery

    # Methods
    def takePhoto(self,count):
        if self.current_battery < 20:
            print("Can't take photo")
        else:
            print(f"Photo clicked by {self.model}")
            self.current_battery -= count
    
```

### Creating Object

```python
apple = phone("iPhone 18", 30, 5000)
samsung = phone("S26 Ultra", 150, 7000)

print(apple.model)
print(apple.current_battery)
print(apple.camera)
print("--------------------------")
print(samsung.model)
print(samsung.current_battery)
print(samsung.camera)
```

### Accessing methods

```python
print(apple.current_battery)
apple.takePhoto(10)
print(apple.current_battery)

print("---------------------")

print(samsung.current_battery)
samsung.takePhoto(30)
print(samsung.current_battery)
```
