### Creating Class

```python
class phone:
    catagory = "Electronics"
    current_battery = 100
    # Constructor
    def __init__(self,model,camera,battery):
        self.model = model
        self.camera = camera
        self.battery = battery

    # Methods
    def takePhoto(self,count):
        if self.current_battery < 20:
            print("Can't take photo")
        else:
            print(f"Photo clicked by {self.model}")
            self.current_battery -= count
```

### Creating another Class

```python
class SmartPhone(phone):
    def __init__(self, model, camera, battery, processor):
        super().__init__(model, camera, battery)
        self.processor = processor

    def fastCharging(self,hour):
        print(f"Fast charging for {hour} hour")


class Cooling_Mechanism():
    def __init__(self,cooling_type):
        self.cooling_type = cooling_type

    def cooling(self):
        print(f"Cooling in progress with {self.cooling_type}")

class pro(SmartPhone, Cooling_Mechanism):
    def __init__(self, model, camera, battery, processor, cooling_type):
        SmartPhone.__init__(self,model, camera, battery, processor)
        Cooling_Mechanism.__init__(self,cooling_type)
```

```python
MacroHard = pro("MMM", 300, 10000, "M10", "Nitrogen")

MacroHard.fastCharging(3)
MacroHard.cooling()
```
