```python
class Camera:
  def __init__(self , name):
    self.name = name

 
  def capture(self):
    print("a photo is captured")
```

```python
class Smart_Phone(Camera):
  def __init__(self,name,resolution):
    super().__init__(name)
    self.resolution = resolution

  #method overriding
  def capture(self):
    print("Photo is captured by a Phone")
```

```python
class DSLR(Camera):
  def __init__(self,name,resolution):
    super().__init__(name)
    self.resolution = resolution

  def capture(self):
    print("Photo is captured  by DSLR")
```

```python
class Drone(Camera):
  def __init__(self,name,resolution):
    super().__init__(name)
    self.resolution = resolution

  def capture(self):
    print("Photo is captured by Drone")
```

```python
phone = Smart_Phone("Phone",30)
dslr = DSLR ("DSLR",200)
drone = Drone("Drone",150)
```

```python
phone.capture()
dslr.capture()
drone.capture()
```
