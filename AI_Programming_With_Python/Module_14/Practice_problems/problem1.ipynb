```python

data = pd.read_csv("sales.csv")
data.head()
```

### Problem 1

```python
sns.lineplot(data=data, x="date", y="sales")
plt.show()
```

### Problem 2

```python
student_data = pd.read_csv("students.csv")
sns.relplot(data=student_data, x= "hours_studied", y= "marks", hue="gender")
```

### Problem 3

```python
student_data = pd.read_csv("students.csv")
sns.relplot(data=student_data, x= "hours_studied", y= "marks", col="subject", kind= "scatter")
plt.show()
```

### Problem 4

```python
student_data = pd.read_csv("students.csv")
sns.histplot(data=student_data, x= "marks", kde= True)
plt.show()
```

### Problem 5

```python
student_data = pd.read_csv("students.csv")
sns.kdeplot(data=student_data, x= "hours_studied")
plt.show()
```
