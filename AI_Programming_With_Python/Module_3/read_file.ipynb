```python
file = open("./test_data.txt","r")

content = file.readlines()

print(content)
```

### Best Practice

```python
with open("./test_data.txt","r") as file:
    for line in file:
        print(line)
```

```python
with open("./test_data.txt","r") as file:
    for line in file:
        print(line, end="")
```
