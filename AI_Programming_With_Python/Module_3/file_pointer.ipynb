```python
with open("./test_data.txt","r") as file:
    print(file.tell()) # Indicates the initial pointer
    print(file.read(5)) # Prints till the index 5
    print(file.tell()) # So, the pointer is now on index 5
    print(file.read()) # Now, it will print from index 6 to end
    print(file.tell()) # Here, the pinter is at the last index
```

```python
with open("./test_data.txt","r") as file:
    file.seek(5) # moves the initial pointer to index 5
    print(file.read()) # starts printing form index 6
```
