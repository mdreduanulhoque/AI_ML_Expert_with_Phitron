```python
## Histogram of price
plt.hist(data=df, x="price", bins=20, edgecolor= "black")
plt.xlabel("Price of House")
plt.ylabel("Frequency")
plt.title("Histogram of price")
plt.show()

## The histogram is right-skewed
```

```python
## Corelation Coefficient
num_df = df.select_dtypes(include=["number"])
corr_matrix = num_df.corr()

sns.heatmap(corr_matrix, annot=True, fmt='0.2f', cmap="coolwarm")
plt.show()

''' "area" and "bathrooms" has the most correlation with price, however stories, parking
and bedroom has some too. '''
```

```python
## Boxplot price grouped by stories
sns.boxplot(x= "price", y= "stories", data = df)
plt.show()
```
